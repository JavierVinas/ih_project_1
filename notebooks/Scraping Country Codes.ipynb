```python
table_list = list(table[0].find_all('td'))
```

```python
table_list
```

```python
# Now we are going to remove all the unnecessary html characters <td> and </td>
```

```python
def replace_html(text):
    text1 = re.sub('<td>', '', str(text))
    text2 = re.sub('</td>', '', str(text1))
    return text2
```

```python
#Paso a lista este string
```

```python
clean_list = list(map(replace_html, table_list))
```

```python
clean_list
```

```python
country = clean_list[0::4]
country_code = clean_list[1::4]
```

```python
country_new = pd.DataFrame(data=[country, country_code],columns=['country', 'country_code'])
```
